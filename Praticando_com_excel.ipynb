```python

df_dup = df.loc[mascara_dup, ['Hora', 'Lote', 'Tipo ', 'Cor', 'Responsavel']]
mascara_dev_nao_x = df['Dev'] != 'x'

df['Produção'] = np.nan
df['Devolução'] = np.nan
df.loc[mascara_dup & dev_x, 'Devolução'] = df['Responsavel']
df.loc[mascara_dup & mascara_dev_nao_x, 'Produção'] = df['Responsavel']
display(df)
```

```python
df['Nomes'] = df[ 'Produção'].fillna('') + df['Devolução'].fillna('')
display(df[['Hora', 'Lote', 'Tipo ','Cor', 'Nomes']])
```

```python
# Identifica as linhas duplicadas
mascara_dup = df.duplicated(subset=['Hora', 'Lote', 'Tipo ', 'Cor'], keep=False)

# Cria a coluna 'Duplicado' marcando as linhas duplicadas
df['Duplicado'] = np.nan
df.loc[mascara_dup, 'Duplicado'] = 'sim'

# Cria a coluna 'Responsaveis' para as linhas duplicadas
df['Responsaveis'] = np.nan

# Preenche a coluna 'Responsaveis' para as linhas duplicadas
for index, row in df[mascara_dup].iterrows():
    hora = row['Hora']
    lote = row['Lote']
    tipo = row['Tipo ']
    cor = row['Cor']

    # Encontra as linhas correspondentes
    linhas_correspondentes = df[(df['Hora'] == hora) &
                                 (df['Lote'] == lote) &
                                 (df['Tipo '] == tipo) &
                                 (df['Cor'] == cor)]

    responsavel_producao = linhas_correspondentes.loc[linhas_correspondentes['Dev'] != 'x', 'Responsavel'].iloc[0] if not linhas_correspondentes.loc[linhas_correspondentes['Dev'] != 'x'].empty else ''
    responsavel_devolucao = linhas_correspondentes.loc[linhas_correspondentes['Dev'] == 'x', 'Responsavel'].iloc[0] if not linhas_correspondentes.loc[linhas_correspondentes['Dev'] == 'x'].empty else ''

    # Combina os nomes e atualiza a coluna 'Responsaveis' nas linhas duplicadas
    df.loc[index, 'Responsaveis'] = f"{responsavel_producao} / {responsavel_devolucao}"

# Exibe o DataFrame com as novas colunas (filtrando apenas as linhas duplicadas, se desejar)
display(df)
```
